```python
#每月跑数

#月份
run_date='202410'
year=run_date[:4]

import pandas as pd

#读取数据
project_info = pd.read_csv('项目信息%s.csv'%run_date)
income_info = pd.read_csv('收入%s.csv'%run_date)
sr_info = pd.read_csv('天翼云科目%s.txt'%year)
corp_list=['深圳','广州','东莞','佛山','中山','惠州','江门','珠海','汕头','揭阳','潮州','汕尾','湛江','茂名','阳江','云浮','清远','肇庆','梅州','韶关','河源']

#统计
income_info=income_info.merge(sr_info.drop_duplicates(),how='left',left_on='收入科目',right_on='天翼云科目')
output1=income_info.groupby('项目编码(18位)',as_index=False).sum('金额')
output2=income_info[income_info['天翼云科目'].notna()].groupby('项目编码(18位)',as_index=False).sum('金额')
income_final=output1.merge(output2,how='left',left_on='项目编码(18位)',right_on='项目编码(18位)')
income_final=income_final.merge(project_info,how='left',left_on='项目编码(18位)',right_on='项目编码(18位)')

#输出
writer = pd.ExcelWriter('含云量统计%s.xlsx'%run_date)
to_excel=pd.DataFrame([],columns=['分公司','DICT项目数','含云项目数','含云率'])
for i in corp_list:
    _ict_cnt=income_final[income_final['分公司']==i].count()
    _ctyun_cnt=income_final[(income_final['分公司']==i)&(income_final['金额_y']>0)].count()
    to_excel.loc[len(to_excel)+1]=[i,_ict_cnt['项目编码(18位)'],_ctyun_cnt['金额_y'],'{:.2%}'.format(_ctyun_cnt['金额_y']/_ict_cnt['项目编码(18位)'])]
_ict_cnt=to_excel.sum()
to_excel.loc[0]=['全省',_ict_cnt['DICT项目数'],_ict_cnt['含云项目数'],'{:.2%}'.format(_ict_cnt['含云项目数']/_ict_cnt['DICT项目数'])]
to_excel.sort_index().to_excel(writer,sheet_name='项目统计',index=False)

to_excel=pd.DataFrame([],columns=['分公司','DICT项目收入','天翼云收入','含云率'])

for i in corp_list:
    _ict_sum=income_final[income_final['分公司']==i].sum()
    to_excel.loc[len(to_excel)+1]=[i,_ict_sum['金额_x'],_ict_sum['金额_y'],'{:.2%}'.format(_ict_sum['金额_y']/_ict_sum['金额_x'])]

_ict_sum=to_excel.sum()
to_excel.loc[0]=['全省',_ict_sum['DICT项目收入'],_ict_sum['天翼云收入'],'{:.2%}'.format(_ict_sum['天翼云收入']/_ict_sum['DICT项目收入'])]
to_excel.sort_index().to_excel(writer,sheet_name='收入统计',index=False)

writer.close()
```
